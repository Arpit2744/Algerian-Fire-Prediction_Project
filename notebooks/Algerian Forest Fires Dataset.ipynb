# Algerian Forest Fires Dataset
#### The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria. 122 instances for each region.

#### The period from June 2012 to September 2012.
#### The dataset includes 11 attributes and 1 output attribute (class)
#### The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classes.

# Additional Variable Information

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations 
2. Temp : temperature noon (temperature max)  in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90 
4. Ws :Wind speed in km/h: 6 to 29 
5. Rain: total day in mm: 0 to 16.8
FWI Components  
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5 
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9 
8. Drought Code (DC) index from the FWI system:  7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5 
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely   â€œFireâ€ and â€œnot Fireâ€

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest.csv')
```

```python
df.head()
```

```python
df.info()
```

## data cleaning

```python
## missing values
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"]=0 # bejaia region 
df.loc[122:,"Region"]=1 # Sidi-bel abbes region
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values 
df=df.dropna().reset_index(drop = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## remove the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.head()
```

```python
df.columns
```

```python
##fic spaces in coumns names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

## change the required col as int data type


```python
df.columns
```

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws',]].astype(int)
```

```python
df.info()
```

#### changing the other columns to float datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('Algerian_cleaned_dataset.csv',index=False)
```
